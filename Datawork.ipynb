Avg Temperature by City

compare cities by its temperature ,see which city is hotter or cooler compares to others

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
sns.barplot(x='city',y='temperature_celsius',data=df)
plt.xticks(rotation=45)
plt.title("Avg Temperature by City")
plt.show()
```

Top 10 Hottest Cities

The graph shows the top 10 hottest cities based on their average temperature

```python
top_cities=df.groupby('city')['temperature_celsius'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,10))
top_cities.plot(kind='bar')
plt.title("Top 10 Hottest Cities")
plt.show()
```

There is very little temperature difference between the 1st and 10th ranked cities.while Dubai is the hottest,Riyadh(the 10th city)still maintains an average temperature well above 30 degree Celsius

Top 10 Coldest Countries

The graph shows the top 10 coldest countries based on their average temperature in celsius

```python
cold10=df.groupby('country')['temperature_celsius'].mean().sort_values().head(10)
plt.figure(figsize=(8,5))
plt.bar(cold10.index,cold10.values)
plt.title("Top 10 Coldest Countries")
plt.xlabel("country")
plt.ylabel("temperature")
plt.xticks(rotation=45)
plt.show()
```

Mongolia is the more coldest and Estonia is the least coldest countries,Estonia and Andorra has only minimal difference and Finland and United States Of America has almostsame second coldest is canada

Monthly Temperature Trend

The graph shows how temperature changes over different month

```python
plt.figure(figsize=(10,10))
sns.lineplot(x='month',y='temperature_celsius',data=df)
plt.title("Monthly Temperature Trend")
plt.show()
```

The warmest month is july(month 7),reaching a peak of approximately 26 degree celsius.The coldest period occures at the beginning of the year,with january(month 1)starting at roughly 16 degree celsius

Temperature Trent Over Time

Showiing how temperature readings(in Celsius) fluctuated over a specific period

```python
plt.figure(figsize=(10,10))
sns.lineplot(x='date',y='temperature_celsius',data=df.head(5000))
plt.xticks(rotation=90)
plt.title("Temperature Trent Over Time")
plt.show()
```

For the majority of the time period (May 16 to June 7),temperature remained constently warm,fluctuating slightly between 24 degree Celsius and 27 degree Celsius,The highest temperature were recorded around June 7,reaching nearly 27 degree Celsius

TEMPERATURE VS LATITUDE

A"Temperature VS Latitude"graph shows the relationship between how far a place is from the equator and its temperature this graph shows that temperature decreases as we move from the equator

```python
sns.scatterplot(x='latitude',y='temperature_celsius',data=df)
plt.title("TEMPERATURE VS LATITUDE")
```

temperature are highest near the equator and decrease significantly towards higher latitude

Temperature Map

This graph shows how temperature is distributed across different location on earth using the geographic position

```python
plt.figure(figsize=(8,5))
plt.scatter(df['longitude'],df['latitude'],c=df['temperature_celsius'])
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("Temperature Map")
plt.colorbar(label='Temperature(degree c)')
plt.show()
```

temperature is strongly influence by latitude,decreasing as we move away from the equator.longitude has minimal impact on temperature variation

Yearly Temperature Trend

This graph illustrate the Yearly Temperature Trend based on the average temperatures

```python
yearly_temp=df.groupby('year')['temperature_celsius'].mean()
yearly_temp.plot(kind='line',marker="o")
plt.xlabel('year')
plt.ylabel('temperature_celsius')
plt.title('Yearly Temperature Trend')
plt.show()
```

the average yearly temperature decreased every year from 2024 to 2026

CORRELATION HEATMAP

visualizes the strength and direction of the linear relationships between three variables(Temperature_celsius,latitude,longitude)

```python
num_df=df[['temperature_celsius','latitude','longitude']]
corr=num_df.corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

A perfect positive relationship(seen on the diagonal comparing a variable to itself),
Temperature vs latitude(-0.38) this shows latitude increases and the temperature tends to decreas(negetive correlation),
Temperature vs longitude(positive correlation)(0.13),
latitude vs londitude(zero correlation)
